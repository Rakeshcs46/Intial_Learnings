```python
#let me inttially copy a pdf  to my directory and then open it 
file = open('Technical Paper.pdf','rb')#rb means reading in binary method since its no longer a textfile 
```

```python
#converting it into pdf file reader object 
pdf_reader = PyPDF3.PdfFileReader(file)
```

```python
pdf_reader.numPages
```
